```python
date_columns = ['AppliedDate','IssuedDate', 'ExpiresDate', 'CompletedDate']

#building_permit_gdf = gpd.read_file("C:/Users/aaron/Fun Mapping Work/Data/Seattle_Building_Permit_Data/Building_Permits_20251209.csv")
#building_permit_gdf = pd.read_csv("C:/Users/aaron/Fun Mapping Work/Data/Seattle_Building_Permit_Data/Building_Permits_20251209.csv", dtype = {'EstProjectCost':float})
building_permit_df = pd.read_csv("C:/Users/aaron/Fun Mapping Work/Data/Seattle_Building_Permit_Data/Building_Permits_20251209.csv",parse_dates = date_columns)
#building_permit_df = pd.read_csv("C:/Users/aaron/Fun Mapping Work/Data/Seattle_Building_Permit_Data/Building_Permits_20251209.csv")
```

```python
## cast EstProjectCost as an int

#building_permit_df['EstProjectCost'] = building_permit_df['EstProjectCost'].str.extract('(\d+)', expand=False)
#building_permit_df['EstProjectCost'] = building_permit_df['EstProjectCost'].astype(int)

building_permit_df['EstProjectCost'] = building_permit_df['EstProjectCost'].str.replace(',', '')
building_permit_df['EstProjectCost'] = pd.to_numeric(building_permit_df['EstProjectCost'], errors='coerce')
building_permit_df = building_permit_df.dropna(subset=['EstProjectCost'])
building_permit_df['EstProjectCost'] = building_permit_df['EstProjectCost'].astype(int)
```
